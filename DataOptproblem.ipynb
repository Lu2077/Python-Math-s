```python
#CARPETA Mi unidad

import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot

df = pd.read_csv("bank(in).csv",sep=";")
#carga el archivo.csv con los datos



df.filter(regex = "age")

promedioedad = df['age'].mean()

print("edad promedio :",promedioedad)

modaedad = pd.Series(df['age'])

print("valor más frecuente :",modaedad.mode()[0])

medianaedad = np.median(df['age'])

print("mediana edad :",medianaedad)
print("----------------------------------")

#edad promedio clientes------------------------------

df.filter(regex = "balance")

promediobalance = df['balance'].mean()

print("balance promedio :",promediobalance)

modabalance = pd.Series(df['balance'])

print("valor más frecuente :",modabalance.mode()[0])

medianabalance = np.median(df['balance'])

print("mediana balance :",medianabalance)
print("----------------------------------")

#blance promedio clientes------------------------------

df.filter(regex = "duration")

promedioduracion = df['duration'].mean()

print("duracion promedio del último contacto en seg :",promedioduracion)

modaduration = pd.Series(df['duration'])

print("valor más frecuente :",modaduration.mode()[0])

medianaduration = np.median(df['duration'])

print("duracion mediana :",medianaduration)
print("----------------------------------")

#duración del último estudio------------------------------

df.filter(regex = "pdays")

promediodias = df['pdays'].mean()

print("media de dias desde la última consulta:",promediodias)

modadias = pd.Series(df['pdays'])

print("valor más frecuente en dias desde la última consulta :"
,modadias.mode()[0])

medianadias = np.median(df['pdays'])

print("dias mediana :",medianadias)
print("----------------------------------")

#días desde la última consulta------------------------------

#EL ARCHIVO bank(in).csv CONTIENE LOS DATOS DE UN BANCO CON RESPECTO
#A UN ESTUDIO BANCARIO DE POTENCIALES CLIENTES O POTENCIALES CLIENTES
#EN ESTE CÓDIGO, SE OBTIENEN LOS PROMEDIOS, LA MODA Y LA MEDIANA DE;
#LA EDAD PROMEDIO, LA EDAD MÁS RECURRENTE DETRO DEL ESTUDIO(MODA) Y LA MEDIANA
#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
#BALANCE PROMEDIO, BALANCE MÁS RECURRENTE DETRO DEL ESTUDIO(MODA) Y LA MEDIANA
#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
#DURACION MEDIA ULTIMA CONSULTA(EN SEGUNDOS),
#DURACION MÁS RECURRENTE DETRO DEL ESTUDIO(MODA) Y LA MEDIANA
#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
#DIAS DESDE LA ULTIMA CONSULTA,
#DIAS MÁS RECURRENTE DETRO DEL ESTUDIO(MODA) Y LA MEDIANA
#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
```
